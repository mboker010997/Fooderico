```python
def checkForValue(row, arr):
    fields_to_check = arr
    for field in fields_to_check:
        if row[field] == 1.0:
            return 1
    return np.nan
```

```python
def markRows(df, new_df, name):
    words = [
        "окрошка", "жарен", "бобы", "горох", "напиток шоколад горячий", "копчен"
    ]
    new_column = []
    for index, row in df.iterrows():
        if any(tok.lower() in row.iloc[0].lower() for tok in words) or new_df[name][index] == 1.0:
            new_column.append(1)
        else:
            new_column.append(np.nan)
    return new_column
```

```python
new_df = df.iloc[:,0]
new_df = createNewColumn(new_df, df, "Мед пчелиный", "Мед")

products_part_4 = {
    "Грибы": mushrooms,
    "Яблоко": apple,
    "Абрикос": apricot,
    "Персик": peach,
    "Киви": kiwi,
}

for product, contained_in in products_part_4.items():
  check_ = partial(checkForValue, arr=contained_in)
  new_df[product] = df.apply(check_, axis=1)

check_wheat = partial(checkForValue, arr=diet4)
new_df["Диета4"] = df.apply(check_wheat, axis=1)
```
